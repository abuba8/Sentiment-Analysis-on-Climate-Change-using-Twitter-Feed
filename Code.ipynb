```python
# making new data frame with dropped NA values 
new_data = data.dropna(axis = 0, how ='any') 
  
# comparing sizes of data frames 
print("Old data frame length:", len(data), "\nNew data frame length:",  
       len(new_data), "\nNumber of rows with at least 1 NA value: ", 
       (len(data)-len(new_data)))

print(new_data.shape)
```

```python
import nltk
nltk.download('punkt')
```

```python
# Import the function to perform stemming
from nltk.stem import PorterStemmer
from nltk import word_tokenize

# Call the stemmer
porter = PorterStemmer()

# Transform the array of tweets to tokens
tokens = [word_tokenize(word) for word in new_data]
# Stem the list of tokens
stemmed_tokens = [[porter.stem(word) for word in new_data] for new_data in tokens] 
# Print the first element of the list
print(stemmed_tokens[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer, ENGLISH_STOP_WORDS, TfidfVectorizer

vect = CountVectorizer(max_features = 100, stop_words = ENGLISH_STOP_WORDS, ngram_range=(1,1))
vect.fit(new_data.tweet)
X = vect.transform(new_data.tweet)

#bigrams + count vectorizer
vect2 = CountVectorizer(max_features = 100, stop_words = ENGLISH_STOP_WORDS, ngram_range=(1,2))
vect2.fit(new_data.tweet)
X2 = vect2.transform(new_data.tweet)
```

```python
X
```

```python
my_array = X.toarray()
X_df = pd.DataFrame(my_array, columns = vect.get_feature_names())

my_array2 = X2.toarray()
X_df2 = pd.DataFrame(my_array2, columns = vect.get_feature_names())
```

```python
from langdetect import detect_langs

languages=[]

for row in range(len(new_data)):
    languages.append(detect_langs(new_data['tweet'].iloc[row]))
    
```

```python
languages = [str(lang).split(':')[0][1:] for lang in languages]
X_df['language'] = languages
X_df2['language'] = languages
```

```python
from nltk import word_tokenize

token = [word_tokenize(item) for item in new_data['tweet']]

print(token)

len_token = []

for i in range(len(token)):
    len_token.append(len(token[i]))
    
X_df['n_words'] = len_token
X_df2['n_words'] = len_token


```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X_df['language'] = labelencoder.fit_transform(X_df.language.values)
X_df2['language'] = labelencoder.fit_transform(X_df2.language.values)
y = labelencoder.fit_transform(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=456)

# Train a logistic regression
clf1 = DecisionTreeClassifier(criterion='entropy').fit(X_train, y_train)
# Predict the labels
y_predicted = clf1.predict(X_test)

# Print accuracy score and confusion matrix on test set
print('Accuracy on the test set: ', accuracy_score(y_test, y_predicted))
print(confusion_matrix(y_test, y_predicted)/len(y_test))
```
